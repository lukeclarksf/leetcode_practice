```python
# LeetCode 1: Two Sum
# Implemented in Python. Returns indices of two numbers that add up to the target.
# Approach: Use a hashmap (dict) to store previously seen values and their indices.

from typing import List, Tuple

def two_sum(nums: List[int], target: int) -> List[int]:
    """Return indices of the two numbers such that they add up to target.

    Uses a dictionary mapping number -> index for O(n) time and O(n) space.
    """
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    raise ValueError("No two sum solution")

# Basic tests to verify behavior
if __name__ == "__main__":
    tests = [
        (([2,7,11,15], 9), [0,1]),
        (([3,2,4], 6), [1,2]),
        (([3,3], 6), [0,1]),
    ]

    for (nums, target), expected in tests:
        result = two_sum(nums, target)
        print(f"nums={nums}, target={target} -> result={result}, expected={expected}")
        assert result == expected

    print("All tests passed!")
```
