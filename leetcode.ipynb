```python
def two_sum(nums: List[int], target: int) -> List[int]:
    """
    LeetCode 1: Two Sum
    Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
    Assumes exactly one solution exists and the same element cannot be used twice.
    """
    seen = {}
    for i, n in enumerate(nums):
        complement = target - n
        if complement in seen:
            return [seen[complement], i]
        seen[n] = i
    return []

tests = [
    (([2, 7, 11, 15], 9), [0, 1]),
    (([3, 2, 4], 6), [1, 2]),
    (([3, 3], 6), [0, 1]),
]

# Run tests (expects tests to be a list of ((nums, target), expected) tuples).
for (nums_input, target_input), expected_output in tests:
    output = two_sum(nums_input, target_input)
    print(f"nums={nums_input}, target={target_input} -> output={output}, expected={expected_output}, pass={output == expected_output}")
```

```python
# LeetCode 2: Add Two Numbers (linked list)
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __repr__(self):
        return f"ListNode({self.val})"

def add_two_numbers(l1: ListNode, l2: ListNode) -> ListNode:
    carry = 0
    dummy = ListNode(0)
    curr = dummy

    while l1 or l2 or carry:
        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0
        s = v1 + v2 + carry
        carry = s // 10
        curr.next = ListNode(s % 10)
        curr = curr.next
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None

    return dummy.next

def list_to_linked(nums):
    dummy = ListNode(0)
    cur = dummy
    for x in nums:
        cur.next = ListNode(x)
        cur = cur.next
    return dummy.next

def linked_to_list(node):
    res = []
    while node:
        res.append(node.val)
        node = node.next
    return res

add_two_tests = [
    (([2, 4, 3], [5, 6, 4]), [7, 0, 8]),
    (([0], [0]), [0]),
    (([9,9,9,9,9,9,9], [9,9,9,9]), [8,9,9,9,0,0,0,1]),
]

for (a, b), expected in add_two_tests:
    l1 = list_to_linked(a)
    l2 = list_to_linked(b)
    out = add_two_numbers(l1, l2)
    out_list = linked_to_list(out)
    print(f"{a} + {b} -> {out_list}, expected={expected}, pass={out_list == expected}")
```

```python
def length_of_longest_substring(s: str) -> int:
    """
    LeetCode 3: Longest Substring Without Repeating Characters
    Returns length of the longest substring without repeating characters using sliding window.
    """
    last_seen = {}
    start = 0
    max_len = 0

    for i, ch in enumerate(s):
        if ch in last_seen and last_seen[ch] >= start:
            start = last_seen[ch] + 1
        last_seen[ch] = i
        max_len = max(max_len, i - start + 1)

    return max_len

# Tests for LeetCode 3
longest_substring_tests = [
    ("abcabcbb", 3),
    ("bbbbb", 1),
    ("pwwkew", 3),
    ("", 0),
    ("au", 2),
    ("dvdf", 3),       # 'vdf'
    ("tmmzuxt", 5),    # 'mzuxt'
]

for s_input, expected in longest_substring_tests:
    out = length_of_longest_substring(s_input)
    print(f"input={s_input!r}, output={out}, expected={expected}, pass={out == expected}")
```

```python
def find_median_sorted_arrays(nums1: List[int], nums2: List[int]) -> float:
    """
    LeetCode 4: Median of Two Sorted Arrays
    Finds the median of two sorted arrays in O(log(min(m, n))) time.
    """
    m, n = len(nums1), len(nums2)
    # Ensure nums1 is the smaller array for binary search
    if m > n:
        return find_median_sorted_arrays(nums2, nums1)

    if n == 0:
        raise ValueError("Both arrays are empty")

    imin, imax, half_len = 0, m, (m + n + 1) // 2

    while imin <= imax:
        i = (imin + imax) // 2
        j = half_len - i

        if i < m and j > 0 and nums2[j - 1] > nums1[i]:
            imin = i + 1  # i is too small, must increase it
        elif i > 0 and j < n and nums1[i - 1] > nums2[j]:
            imax = i - 1  # i is too big, must decrease it
        else:
            # Correct partition found
            if i == 0:
                max_left = nums2[j - 1]
            elif j == 0:
                max_left = nums1[i - 1]
            else:
                max_left = max(nums1[i - 1], nums2[j - 1])

            if (m + n) % 2 == 1:
                return float(max_left)

            if i == m:
                min_right = nums2[j]
            elif j == n:
                min_right = nums1[i]
            else:
                min_right = min(nums1[i], nums2[j])

            return (max_left + min_right) / 2.0

    # If we get here, input arrays were not valid sorted arrays
    raise ValueError("Input arrays are not sorted or invalid for median computation")


# Tests for LeetCode 4
median_tests = [
    (([1, 3], [2]), 2.0),
    (([1, 2], [3, 4]), 2.5),
    (([0, 0], [0, 0]), 0.0),
    (([], [1]), 1.0),
    (([2], []), 2.0),
    (([1, 2, 3], [4, 5, 6]), 3.5),
    (([1, 3, 8], [7, 9, 10, 11]), 8.0),
    # Use existing notebook variables a and b (both lists of 9s)
    ((a, b), 9.0),
]

for (n1, n2), expected_median in median_tests:
    result_median = find_median_sorted_arrays(n1, n2)
    print(f"{n1} & {n2} -> median={result_median}, expected={expected_median}, pass={result_median == expected_median}")
```

```python
def longest_palindrome(s: str) -> str:
    """
    LeetCode 5: Longest Palindromic Substring
    Expand around center approach (O(n^2) time, O(1) space).
    Returns the longest palindromic substring.
    """
    if not s:
        return ""

    def expand_around_center(left: int, right: int) -> int:
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        # length = right - left - 1 after overshoot
        return right - left - 1

    start, end = 0, 0
    for i in range(len(s)):
        len1 = expand_around_center(i, i)     # odd length
        len2 = expand_around_center(i, i + 1) # even length
        cur_len = max(len1, len2)
        if cur_len > (end - start + 1):
            start = i - (cur_len - 1) // 2
            end = i + cur_len // 2

    return s[start:end + 1]


# Tests for LeetCode 5
palindrome_tests = [
    ("babad", ["bab", "aba"]),   # both "bab" and "aba" are acceptable
    ("cbbd", ["bb"]),
    ("a", ["a"]),
    ("ac", ["a", "c"]),
    ("", [""]),
    ("bb", ["bb"]),
    ("forgeeksskeegfor", ["geeksskeeg"]),
    ("tmmzuxt", ["mm"]),  # uses s_input-like string existing in notebook
]

for s_test, expected_list in palindrome_tests:
    out = longest_palindrome(s_test)
    print(f"input={s_test!r}, output={out!r}, expected one of {expected_list}, pass={out in expected_list}")
```

```python
def convert(s: str, numRows: int) -> str:
    """
    LeetCode 6: ZigZag Conversion
    Convert string s into zigzag pattern on numRows and then read line by line.
    """
    if numRows == 1 or numRows >= len(s):
        return s

    rows = [''] * numRows
    cur_row = 0
    going_down = False

    for ch in s:
        rows[cur_row] += ch
        if cur_row == 0 or cur_row == numRows - 1:
            going_down = not going_down
        cur_row += 1 if going_down else -1

    return ''.join(rows)


# Tests for LeetCode 6
zigzag_tests = [
    (("PAYPALISHIRING", 3), "PAHNAPLSIIGYIR"),
    (("PAYPALISHIRING", 4), "PINALSIGYAHRPI"),
    (("A", 1), "A"),
    (("AB", 1), "AB"),
    (("ABCDE", 4), "ABCED"),
]

for (s_test, rows_test), expected in zigzag_tests:
    out = convert(s_test, rows_test)
    print(f"s={s_test!r}, numRows={rows_test} -> out={out!r}, expected={expected!r}, pass={out == expected}")
```

```python
def reverse(x: int) -> int:
    """
    LeetCode 7: Reverse Integer
    Reverse digits of a 32-bit signed integer. Return 0 on overflow.
    """
    INT_MAX = 2**31 - 1
    INT_MIN = -2**31

    sign = -1 if x < 0 else 1
    x_abs = abs(x)
    rev = 0

    # target is the absolute limit for rev (positive)
    target = INT_MAX if sign == 1 else 2**31

    while x_abs:
        pop = x_abs % 10
        x_abs //= 10
        # Overflow check before pushing digit
        if rev > target // 10 or (rev == target // 10 and pop > target % 10):
            return 0
        rev = rev * 10 + pop

    rev_signed = sign * rev
    if rev_signed < INT_MIN or rev_signed > INT_MAX:
        return 0
    return rev_signed


# Tests for LeetCode 7
reverse_tests = [
    (123, 321),
    (-123, -321),
    (120, 21),
    (0, 0),
    (1534236469, 0),       # will overflow -> 0
    (-1534236469, 0),      # will overflow -> 0
    (1463847412, 2147483641),  # no overflow
]

for x_input, expected_output in reverse_tests:
    out = reverse(x_input)
    print(f"x={x_input} -> out={out}, expected={expected_output}, pass={out == expected_output}")
```
