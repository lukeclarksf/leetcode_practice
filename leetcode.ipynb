```python
def two_sum(nums: List[int], target: int) -> List[int]:
    """
    LeetCode 1: Two Sum
    Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
    Assumes exactly one solution exists and the same element cannot be used twice.
    """
    seen = {}
    for i, n in enumerate(nums):
        complement = target - n
        if complement in seen:
            return [seen[complement], i]
        seen[n] = i
    return []

tests = [
    (([2, 7, 11, 15], 9), [0, 1]),
    (([3, 2, 4], 6), [1, 2]),
    (([3, 3], 6), [0, 1]),
]

# Run tests (expects tests to be a list of ((nums, target), expected) tuples).
for (nums_input, target_input), expected_output in tests:
    output = two_sum(nums_input, target_input)
    print(f"nums={nums_input}, target={target_input} -> output={output}, expected={expected_output}, pass={output == expected_output}")
```

```python
# LeetCode 2: Add Two Numbers (linked list)
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __repr__(self):
        return f"ListNode({self.val})"

def add_two_numbers(l1: ListNode, l2: ListNode) -> ListNode:
    carry = 0
    dummy = ListNode(0)
    curr = dummy

    while l1 or l2 or carry:
        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0
        s = v1 + v2 + carry
        carry = s // 10
        curr.next = ListNode(s % 10)
        curr = curr.next
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None

    return dummy.next

def list_to_linked(nums):
    dummy = ListNode(0)
    cur = dummy
    for x in nums:
        cur.next = ListNode(x)
        cur = cur.next
    return dummy.next

def linked_to_list(node):
    res = []
    while node:
        res.append(node.val)
        node = node.next
    return res

add_two_tests = [
    (([2, 4, 3], [5, 6, 4]), [7, 0, 8]),
    (([0], [0]), [0]),
    (([9,9,9,9,9,9,9], [9,9,9,9]), [8,9,9,9,0,0,0,1]),
]

for (a, b), expected in add_two_tests:
    l1 = list_to_linked(a)
    l2 = list_to_linked(b)
    out = add_two_numbers(l1, l2)
    out_list = linked_to_list(out)
    print(f"{a} + {b} -> {out_list}, expected={expected}, pass={out_list == expected}")
```

```python
def length_of_longest_substring(s: str) -> int:
    """
    LeetCode 3: Longest Substring Without Repeating Characters
    Returns length of the longest substring without repeating characters using sliding window.
    """
    last_seen = {}
    start = 0
    max_len = 0

    for i, ch in enumerate(s):
        if ch in last_seen and last_seen[ch] >= start:
            start = last_seen[ch] + 1
        last_seen[ch] = i
        max_len = max(max_len, i - start + 1)

    return max_len

# Tests for LeetCode 3
longest_substring_tests = [
    ("abcabcbb", 3),
    ("bbbbb", 1),
    ("pwwkew", 3),
    ("", 0),
    ("au", 2),
    ("dvdf", 3),       # 'vdf'
    ("tmmzuxt", 5),    # 'mzuxt'
]

for s_input, expected in longest_substring_tests:
    out = length_of_longest_substring(s_input)
    print(f"input={s_input!r}, output={out}, expected={expected}, pass={out == expected}")
```
